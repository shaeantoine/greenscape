```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import xgboost as xgb
from pathlib import Path
import os
```

```python
aapl = 'stock-price-predictions/price/AAPL.csv'
df = pd.read_csv(aapl)
```

```python
df.drop(['High', 'Low'], axis=1, inplace=True)
```

```python
df['Percentage Change'] = df['Close'].pct_change()
```

```python
df['Up'] = np.where(df['Percentage Change'] > 0, 1, 0)
```

```python
plt.figure(figsize=(12, 6))

plt.plot(df['Date'], df['Percentage Change'])
plt.title('Percentage Change in Stock Price')

plt.xlabel('Date')
plt.ylabel('Percentage Change')
```
